## Pandas Demo

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams['font.size'] = 22
```

## We will see just how easy it makes our lives!

```python
covid_df = pd.read_csv("Covid_Data.csv")

covid_df

# print(covid_df['date'])
```

```python
#custom x-axis each index corresponds to a new month
locations = [1, 33, 62,93,123,154,184,215,247]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Oct']

f, (ax1,ax2) = plt.subplots(1,2, figsize=(20,8))

ax1.set_title("New Cases and New Deaths Throughout the Year")
ax1.plot(covid_df['new_cases_smoothed'], 'b-')
ax1.plot(covid_df['new_cases'], 'b.', label="New Cases")
ax1.plot(covid_df['new_deaths_smoothed'], 'r-')
ax1.plot(covid_df['new_deaths'], 'r.', label= "New Deaths")
ax1.set_xticks(locations)
ax1.set_xticklabels(months)
ax1.legend()
ax1.grid()

ax2.set_title("Total Cases and Total Deaths Throughout the Year")
ax2.plot(covid_df['total_cases'], 'b.', label="Total Cases")
ax2.plot(covid_df['total_deaths'], 'r.', label= "Total Deaths")
ax2.set_xticks(locations)
ax2.set_xticklabels(months)
ax2.legend()
ax2.grid()

f.tight_layout()
plt.show()
```

## Now lets create and store our own files with pandas!

```python
# new_dataframe = pd.DataFrame(Dictionary)

# We need to make a dictionary

dates_dict = {'Months':months, 'Indeces':locations}

# print(dates_dict)

date_locations_df = pd.DataFrame(dates_dict)

date_locations_df
```

```python
#Let's Save the DataFrame!

date_locations_df.to_csv('date_locations.csv')
```
