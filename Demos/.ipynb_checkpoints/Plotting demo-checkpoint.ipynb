# Importing Some Cool Data!

```python
import numpy as np
```

#### Loading data:

```python
data = np.loadtxt('dogecoin.txt') #Values of DogeCoin in USD over time
```

```python
print(np.shape(data))
print("Mean: ", np.mean(data))
print("Standard Deviation: ", np.std(data))
print("Max: ", np.max(data))
print("Min: ", np.min(data))
```

# Let's learn how to plot!!!


#### Don't forget these:

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Now we plot! Start by creating a figure, then set up your axes and make it pretty:

```python
plt.figure(figsize=(12,8))
indices = list(range(len(data)))
plt.plot(indices,data)
plt.xlim(1000,1500)

plt.title("DOGECOIN",fontsize=30)
plt.ylabel("USD",fontsize=22)
plt.xlabel("Days Since Nov. 9 2017",fontsize=22)
plt.show()
```
