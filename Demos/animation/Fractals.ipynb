```python
import numpy as np
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.figsize'] = [10, 10]
```

```python
def mandelbrot(z, c):
    z_new = z**2 + c
    return z_new

def is_div(z, c):
    for n in range(n_max):
        if abs(z) > n_max:
            return n
        z = mandelbrot(z, c)
    return n_max
```

```python
# Mandelbrot set
n = 501 # higher num increases resolution
n_max = 50  # number of iteration
x_min, x_max = -2, 0.5
y_min, y_max = -1.25, 1.25

matrix = []
x = []
for i in range(n):
    i = x_min + i * (x_max - x_min) / (n - 1)
    x += [i]
y = []
for i in range(n):
    i = y_min + i * (y_max - y_min) / (n - 1)
    y += [i]

for ix in x:
    column = []
    for iy in y:
        c = complex(ix, iy)
        n = is_div(0, c)
        column.append(n)
    matrix.append(column)

plt.imshow(matrix)
```

```python
# Julia set
c = complex(-0.80, 0.17) # change this value for new fractal
n = 701
n_max = 70
x_min, x_max = -1.5, 1.5
y_min, y_max = -1.5, 1.5

matrix = []
x = []
for i in range(n):
    i = x_min + i * (x_max - x_min) / (n - 1)
    x += [i]
y = []
for i in range(n):
    i = y_min + i * (y_max - y_min) / (n - 1)
    y += [i]

for ix in x:
    column = []
    for iy in y:
        zo = complex(ix, iy)
        n = is_div(zo, c)
        column.append(n)
    matrix.append(column)
    
plt.imshow(matrix, cmap = 'bone')  # change cmap for new colors
```
