## Astropy Tutorial and More 

Below, I will walk you through the astropy tutorial "Viewing and manipulating FITS images"

```python
#Importing image data from the given website
#NOTE: a .fits file is a Flexible Image Transport System--this is the main format for images in astronomy
#you receive a fits file from a telescope when you observe
#you can also create or modify fits files and there are many tutorials for this online
image_file = download_file('http://data.astropy.org/tutorials/FITS-images/HorseHead.fits', cache=True )

#opening fits image and printing out its information
#the information is used to determine which channel (listed as No.) contains the data--here we want PRIMARY, so channel 0
hdu_list = fits.open(image_file)
hdu_list.info()

#assigning the image data in channel 0 to a variable
image_data = hdu_list[0].data
#printing the associated header
#note a header contains information about the image--which telescope took it, the pixel size, etc.
print(hdu_list[0].header)

print(type(image_data))
print(image_data.shape)
hdu_list.close()
```

```python
#another command for extracting data from a fits file
image_data = fits.getdata(image_file)
#printing the shape of the image i.e. how many pixels are on each image axis
print(image_data.shape)

#plotting the image with a colomap of afmhot
plt.imshow(image_data, cmap='afmhot')
plt.title('Horsehead Nebula', fontsize = 25)
plt.colorbar()

# To see more color maps
# https://matplotlib.org/tutorials/colors/colormaps.html
```

```python
#printing some statistics associated with the image
print('Min:', np.min(image_data))
print('Max:', np.max(image_data))
print('Mean:', np.mean(image_data))
print('Stdev:', np.std(image_data))
```

## Visualising Binning

Homework 2 asked you to solve a problem which involved a common concept in astronomy called 'binning.' The three figures below will help you visualize the influence binning has on how we view a spectrum.

Binning is the process in which adjacent pixels are combined to improve the signal to noise (S/N) in an image or spectrum. For a spectrum, a bin number of 1000 (meaning we are using 1000 bins) would mean that we simply take the range presented on the x-axis and divide it into 1000 equally spaced regions. In these regions, every count is summed to get 'total counts' within this bin. Plotting each bin as a historgram then gives you a spectrum.

You will see below that the number of bins you using for binning is $\textbf{REALLY}$ important. If we use too few bins, then our resolution (ability to distinguish two peaks from one preak) goes down. If we use too many bins, then we will find that the spectrum gets very 'hairy' and our amplitude (the tallest point in the peak) goes way down.

```python
'''Visualising Binning'''

### Here we are using 1000 bins
NBINS = 1000
plt.hist(image_data.flatten(), NBINS)
plt.xlabel('Wavelength (Angstrom)')
plt.ylabel('Counts')
plt.show()
```

```python
## We are now using 10 bins --notice that we cannot prove there are two separate peaks here
NBINS = 10
plt.hist(image_data.flatten(), NBINS)
plt.xlabel('Wavelength (Angstrom)')
plt.ylabel('Counts')
plt.show()
```

```python
## Now we are using 10000 bins -- notice that the tallest point in each peak has less counts than when we used 1000 bins
NBINS = 10000
plt.hist(image_data.flatten(), NBINS)
plt.xlabel('Wavelength (Angstrom)')
plt.ylabel('Counts')
plt.show()
```

## Uploading data from your computer 

Below is the ROSAT All-Sky X-ray Survery image at 3/4 keV. This example is to show you how to upload data directly from your computer as opposed to downloading it from the internet as the astropy tutorial had you do.

A couple interesting notes:

$\textbf{What is meant by 'image at 3/4 keV'?}$

The first thing to note here, is that X-ray astronomers use energies instead of wavelength or frequency to denote what kind of photons they are looking at. The keV above is just short for kilo-electronvolts.

In X-ray astronomy (and radio and others), there are specific 'bands' which we observe in. This is because the range of wavelengths X-ray and Radio cover is much larger than the range of wavelengths optical covers. Therefore, it is unrealistic to create a $\textbf{good}$ (good is subjective to the research you want to complete) detector that can view the ENTIRE X-ray or ENTIRE radio range at once. Therefore, the 3/4 keV band shown below is just denoting the range of wavelengths observed in this image. For example, there is also a 1/4 keV X-ray band that looks at lower energy photons. 


$\textbf{What are the black streaks in the image?}$

ROSAT was an X-ray satellite that orbited around the Earth. All satellites (or instruments in general) have a field-of-view (FOV). The FOV simply tells you how much of the sky can be seen at once. ROSAT had two different instruments, the PSPC which had a FOV = 2 degrees (reasonably sized) and the HRI which had a FOV = 40 arcmin (pretty small). Different sizes of FOV are preferable for different kinds of missions--meaning for a project where you want to observe the whole sky you will want a large FOV and for a project where you want to look at something small and specific (i.e. supernovae, filaments, etc.) you will want a small FOV.

The black streaks are caused by a lack of data. This means that ROSAT's FOV never overlapped with those areas on the sky, and therefore we do not have any data for these lines so they appear black. A cool thing these lines do show is how ROSAT observed across the sky.

```python
'''Uploading an image from your computer'''
#repeating the steps above for an image that you already have downloaded

hdu_list1 = fits.open('rass_m.fits')
image_data1 = hdu_list1[0].data

plt.imshow(image_data1, cmap='afmhot', vmin=0, vmax=1e3)
plt.title('ROSAT All-Sky X-Ray Survey (3/4 keV)', fontsize = 25)
plt.colorbar(label = 'Energy (keV)')

print('Min:', np.min(image_data1))
print('Max:', np.max(image_data1))
print('Mean:', np.mean(image_data1))
print('Stdev:', np.std(image_data1))
```
