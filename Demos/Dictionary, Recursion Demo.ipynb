# Dictionaries

```python
def factorial(x):
    product = 1
    for i in range(1, x+1):
        product = product * i
    return product

def while_factorial(x):
    if x == 0:
        return 1
    else:
        product = 1
        i = 1
        while i <= x:
            product = product * i
            i += 1
        return product

factorial(3)
while_factorial(0)
```

```python
planet_day_in_hours = [('Earth', 24), ('Mercury', 1416), ('Venus', 5832), ('Mars', 24.6), 
                        ('Jupiter', 9.8), ('Saturn', 10.2), ('Uranus', 17.2), ('Neptune', 16.1), ('Pluto', 154)]
```

```python
# create a dictionary
planet_days = {} # planet_days = dict()

# add an entry for how many hours in day on Earth
planet_days["Earth"] = 24
planet_days["Mercury"] = 1416
```

```python
planet_days
```

```python
# alternative way: create the dictionary from planet_day_in_hours 
planet_days = dict(planet_day_in_hours)
planet_days
```

```python
# write code to compare ratio of Earth day to other planet days
for planet in planet_days.keys():
    planet_days[planet] = planet_days[planet] / 24
planet_days
```

```python
# Delete the (key, value) entry for Pluto because Pluto is not a planet :(
del planet_days["Pluto"]
planet_days
```

# Recursion

```python
# 0, 1, 1, 2, 3, 5, 8, 13...

def fib_while(n):
    '''
    While loop version of fibonacci sequence
    Returns the nth Fibonacci number
    '''
    fib1 = 0
    fib2 = 1
    k = 1
    while k < n:
        curr = fib1 + fib2
        fib1 = fib2
        fib2 = curr
        k = k + 1
    return curr

def fib_rec(n):
    '''
    Recursion version of fibonacci sequence
    Returns the nth Fibonacci number
    '''
    if n == 0:
        return 0 # base case
    if n == 1:
        return 1 # base case
    return fib_rec(n-2) + fib_rec(n-1) # recursive call

print(fib_while(10))
print(fib_rec(10))
```

```python
from math import sqrt

def sqrt_6_plus(x):
    '''
    Computes the sqrt of 6 + ... x times
    where x >= 0
    '''
    if x == 0: # base case 
        return 0
    
    if x == 1: # base case
        return sqrt(6) 
    
    return sqrt(6 + sqrt_6_plus(x - 1)) # recursive call

sqrt_6_plus(1000)
```
