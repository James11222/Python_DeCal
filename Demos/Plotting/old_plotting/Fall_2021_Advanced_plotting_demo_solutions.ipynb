## Starting Simple

For the first exercise we will simply plot 3D data using `plt.imshow`. Let's use what we just learned on the homework, a checkerboard!

```python
#create a 2D checkerboard array using the code we wrote in HW3
#checkerboard = ...

checkerboard = np.ones((8,8), dtype = int)
checkerboard[1::2,::2] = 0
checkerboard[::2,1::2] = 0
print(checkerboard)
```

```python
#plot checkerboard

plt.figure(figsize=(6,6))
plt.imshow(checkerboard,cmap='hot')
plt.xticks([])
plt.yticks([])
plt.show()
```

## Plotting Continuous 3D Functions

let's plot a function $f(x,y)$ that we may come across in multivariable calculus for example

```python
# create x and y grid for our function

# Generate the x and y values for a 2D grid
x, y = np.meshgrid(np.linspace(-10, 10, 1000),
                   np.linspace(-10, 10, 1000))

# write multivariable function

f = np.sin(x*y) + np.cos(x)

# plot using imshow()

plt.figure(figsize=(8,6))
plt.imshow(f)             # Plot the 2D array
plt.set_cmap('inferno')   # Set the color scheme ("jet" is matplotlib's default)
plt.colorbar()            # Display a legend for the color values
plt.show()
```

## Making Subplots

```python
# make x, y1, y2

x = np.linspace(0,2*np.pi,500)
y1 = np.sin(x)
y2 = np.sin(x) + np.random.uniform(-.5,.5,500)

# initialize figure

f, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4))

# fill figure with subplots

ax1.plot(x, y1, 'r-')
ax1.set_title("sin(x) with no noise")

ax2.plot(x,y2, 'r-')
ax2.plot(x, y1, 'k-')
ax2.set_title("sin(x) with noise")

plt.show()
```
