```python
import numpy as np
# Set up matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (20,13)
from astropy.io import fits
import matplotlib.transforms as trans
```

# Venus

```python
hdu_list = fits.open('venus.fits')
hdu_list.info()


#assigning the image data in channel 0 to a variable
image_data = hdu_list[0].data
#printing the associated header
#note a header contains information about the image--which telescope took it, the pixel size, etc.
#print(hdu_list[0].header)

#print(type(image_data))
#print(image_data.shape)
hdu_list.close()
```

```python
plt.imshow(image_data, cmap='inferno', vmax=5000)
plt.colorbar()
# plt.ylim(450,550)
# plt.xlim(850,950)
```

# Eagle Nebula

```python
image_data = fits.getdata('Eagle_Nebula.fits')

print(type(image_data))
print(image_data.shape)
```

```python
"""Plotting Raw Data"""
plt.figure()
plt.imshow(image_data, cmap='inferno',vmin=0, vmax=400) #If the image looks like nothing, lower max pixel counts
plt.colorbar()
plt.show()

import scipy.misc
from scipy import ndimage

# define transformed line
rotated_img = ndimage.rotate(image_data, 270)

"""Rotating + Reshaping Graph to look nice"""
plt.figure()
plt.imshow(rotated_img, cmap='inferno',vmin=0, vmax=400)
plt.colorbar()
plt.xlim(1600,0)
plt.show()

# To see more color maps
# http://wiki.scipy.org/Cookbook/Matplotlib/Show_colormaps
```

# Sombrero Galaxy

```python
image_data = fits.getdata('sombrero.fits')

print(type(image_data))
print(image_data.shape)
print(image_data)
```

```python
# """Plotting Raw Data"""
# plt.figure()
# plt.imshow(image_data, cmap='inferno') #If the image looks like nothing, lower max pixel counts
# plt.colorbar()
# plt.show()

import scipy.misc
from scipy import ndimage

# rotate an image using scipy
rotated_img = ndimage.rotate(image_data, 180)

"""Rotating + Reshaping Graph to look nice"""
plt.figure()
plt.imshow(rotated_img, cmap='inferno', vmin = 6000, vmax = 21000)
plt.colorbar()
plt.show()

# To see more color maps
# http://wiki.scipy.org/Cookbook/Matplotlib/Show_colormaps
```
