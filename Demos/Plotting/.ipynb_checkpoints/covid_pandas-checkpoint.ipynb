```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['mathtext.fontset'] = 'stix'
plt.rcParams['font.family'] = 'STIXGeneral'
plt.rcParams['font.size'] = 22
```

```python
covid_df = pd.read_csv('Covid_Data.csv')

covid_df

# covid_df['total_cases']
```

```python
months=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'Jul', 'Aug', 'Oct']
locations = [1,33,62,93,123,154,184,215,247]

f, (ax1,ax2) = plt.subplots(1,2, figsize=(20,6))

ax1.set_title("New Cases and Deaths in 2020")
ax1.plot(covid_df['new_cases'], 'b.')
ax1.plot(covid_df['new_cases_smoothed'], 'b-')
ax1.plot(covid_df['new_deaths'], 'r.')
ax1.plot(covid_df['new_deaths_smoothed'], 'r-')
ax1.set_xticks(locations)
ax1.set_xticklabels(months)
ax1.grid()

ax2.set_title("Total Cases and Deaths in 2020")
ax2.plot(covid_df['total_cases'], 'b.')
# ax2.plot(covid_df['new_cases_smoothed'], 'b-')
ax2.plot(covid_df['total_deaths'], 'r.')
# ax2.plot(covid_df['new_deaths_smoothed'], 'r-')
ax2.set_xticks(locations)
ax2.set_xticklabels(months)
ax2.grid()

f.tight_layout()
```
