```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import colors
```

```python
x, y = np.meshgrid(np.linspace(-10,10,1000),
                  np.linspace(-10,10,1000))

f = np.sin(x) + np.sin(-y)

plt.figure(figsize=(12,12))
plt.imshow(f, cmap = 'jet')
plt.colorbar()
plt.show()
```

```python
data = np.loadtxt('2d_data.csv', delimiter=',', unpack=False)
```

```python
custom_cmap = colors.ListedColormap(['limegreen', 'black'])

plt.figure(figsize=(12,12))
plt.imshow(data, cmap = custom_cmap)
plt.show()
```

```python
from astropy.io import fits
from scipy import ndimage

image_data = fits.getdata('Plotting_FITS_Files/sombrero.fits')
rotated_img = ndimage.rotate(image_data,180)
plt.figure(figsize=(20,13))
plt.imshow(rotated_img, cmap = 'inferno', vmin = 6000, vmax =21000)
plt.colorbar()
plt.show()
```

```python
image_data = fits.getdata('Plotting_FITS_Files/Eagle_Nebula.fits')
rotated_img = ndimage.rotate(image_data,270)
plt.figure(figsize=(20,13))
plt.imshow(rotated_img, cmap = 'inferno', vmin = 0, vmax =400)
plt.colorbar()
plt.show()
```
