```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lam_0 = .5
a_0 = 5
PSI = lambda x,lam,a: np.sqrt(lam/np.pi)*np.exp(-lam * (x - a)**2)
x = np.linspace(0,10, 1000)
```

```python
plt.plot(x, PSI(x,lam_0,a_0), label="$\\rho(x)$")
plt.title("$\\rho(x) vs. x$")
plt.legend()
#plt.savefig('137A_1.eps', format='eps', dpi=300, bbox_inches='tight')
plt.show()
```

```python
x = np.linspace(-5,5, 1000)
plt.figure(figsize=(10,6))
plt.plot(x, PSI(x,1,0), label="$\\lambda = 1$")
plt.plot(x, PSI(x,10,0), label="$\\lambda = 10$")
plt.plot(x, PSI(x,100,0), label="$\\lambda = 100$")
plt.xlim(-5,5)
plt.legend()
#plt.savefig('137A.eps', format='eps', dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(x, PSI(x,10000000,0), label="$\\lambda = 10000000$")
plt.xlim(-5,5)
plt.legend()
plt.show()
```

```python
b = 5
y = np.heaviside(x - b, 0.5)
plt.plot(x,y)
plt.title("$\\Theta(x - b)$ vs. x")
#plt.savefig('137A_2.eps', format='eps', dpi=300, bbox_inches='tight')
plt.show()
```

```python
G = 6.67e-11 #kg^-1 m^3 s^-2
pi = 3.14
def Luminosity(radius, temp):
    sigma_sb = 5.67e-8 #watt m^-2 K^-4
    L = 4*pi*radius**2 * sigma_sb * temp**4
    return L
```
