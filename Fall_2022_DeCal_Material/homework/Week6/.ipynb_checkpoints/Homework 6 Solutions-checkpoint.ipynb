# Numerical Differentiation

SciPy offers a library function to compute derivatives. It uses a central difference formula, but with additional ability to use more than two points. Here is an example of its use, it is called `scipy.mis.derivative`. See full documentation at http://docs.scipy.org/doc/scipy/reference/generated/scipy.misc.derivative.html

```python
#Code Preamble#
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from scipy.misc import derivative
import time

# Compute the derivative of cos(x) at x = pi
t0 = time.time()
dy_scipy_1 = derivative(func=np.cos, x0=np.pi, dx=1e-6, n=1, order=3)
t1 = time.time()
dy_scipy_2 = derivative(func=np.cos, x0=np.pi, dx=1e-6, n=1, order=101)
t2 = time.time()

dt1 = t1 - t0
dt2 = t2 - t1
print("Derivative using order = 3: %.5f, takes %.5f sec to compute using your computer" % (dy_scipy_1, dt1))
print("Derivative using order = 101: %.5f, takes %.5f sec to compute using your computer" % (dy_scipy_2, dt2))
```

The arguments here are as follows:

    func:  the name of the function whose derivative you want to calculate
    x0:    the location of the value (scalar) or values (numpy array) where you want to evaluate the derivative
    dx:    spacing between the points it generates (by evaluating `func`) to calculate the differences
    n:     the number of derivatives. `n=1` means first derivative of `func`
    order: number of points to use in the difference method. Must be odd.

```python
def centered_dy(y, x):
    '''
        Uses centered differences (see below) to estimate the derivatives at each value of x, 
        except for the first and last values. The derivative at the first value of x is estimated 
        using a forward difference. The derivative at the last value of x is estimated 
        using a backward difference.
            dy/dx at x[i] is approximated by (y[i+1] - y[i-1]) / (x[i+1]-x[i-1])
    '''
    dyc = [0.0]*len(x)
    dyc[0] = (y[0] - y[1])/(x[0] - x[1])
    for i in range(1,len(y)-1):
        dyc[i] = (y[i+1] - y[i-1])/(x[i+1]-x[i-1])
    dyc[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])

    return dyc

###Your Code Here###


dy_c = centered_dy(np.cos(x),x)
dy_analytic = -np.sin(x)

diff_c = dy_c - dy_analytic


x = np.linspace(0.0,2*np.pi,100)
function = np.cos(x)

dy_sp = derivative(func=np.cos, x0=x, dx=1e-6, n=1, order=3)
dy_analytical = -np.sin(x)
diff = dy_sp - dy_analytical

plt.figure(figsize=(8,6))
plt.title("The Differences of the dy_c and misc.derivative with analytical")
plt.plot(x,diff,'r-',label="scipy diff")
plt.plot(x,diff_c,'k-',label="centered diff")
plt.legend()
plt.show()

plt.figure(figsize=(8,6))
plt.title("N Derivatives")
for N in range(0,4):
    x = np.linspace(0.0,2*np.pi,100)
    dy_sp = derivative(func=np.cos, x0=x, dx=1e-5, n=N, order=11)
    dy_analytical = -np.sin(x)
    plt.plot(x,dy_sp,label=N)
plt.legend()
plt.show()

print("""The misc.derivative function of scipy breaks down after 2 derivatives unless we
decide to increase the dx value from 1e-6 to 1e-5. This is something we have to note if we
want to utilize the derivative function from scipy. It begins to break down as n begins to 
increase. In order to counteract this we have to increase dx size. if we change the dx to 1e-4 or 1e-3 
the 3rd derivative will appear correctly. We can also see that the order has a small effect 
on this as well if we increase order points then it can help graph the nth derivative.""")
```

```python
###Your Code Here###

import scipy as sp
import scipy.integrate


#defining all our functions to integrate
def sin(x):
    y = np.sin(x)
    return y

def poly(x):
    y = x**2 +x + 1
    return y 

#part 1
#we follow the same steps for each part, do integral for trapz, quad, and simps with varying step sizes
#note quad doesn't change with step size because it is a fixed step size that it custom chooses.
integral_quad_1, error_quad_1 = scipy.integrate.quad(sin, 0, np.pi)
x_1 = np.linspace(0,np.pi,100)
integral_trapz_1 = scipy.integrate.trapz(sin(x_1), x_1)
integral_simps_1 = scipy.integrate.simps(sin(x_1), x_1)
x_1_2 = np.linspace(0,np.pi,25)
integral_trapz_1_2 = scipy.integrate.trapz(sin(x_1_2), x_1_2)
integral_simps_1_2 = scipy.integrate.simps(sin(x_1_2), x_1_2)
x_1_3 = np.linspace(0,np.pi,50)
integral_trapz_1_3 = scipy.integrate.trapz(sin(x_1_3), x_1_3)
integral_simps_1_3 = scipy.integrate.simps(sin(x_1_3), x_1_3)


#part 2
integral_quad_2, error_quad_2 = scipy.integrate.quad(poly, 2, 4)
x_2 = np.linspace(2,4,100)
integral_trapz_2 = scipy.integrate.trapz(poly(x_2), x_2)
integral_simps_2 = scipy.integrate.simps(poly(x_2), x_2)
x_2_2 = np.linspace(2,4,25)
integral_trapz_2_2 = scipy.integrate.trapz(poly(x_2_2), x_2_2)
integral_simps_2_2 = scipy.integrate.simps(poly(x_2_2), x_2_2)
x_2_3 = np.linspace(2,4,50)
integral_trapz_2_3 = scipy.integrate.trapz(poly(x_2_3), x_2_3)
integral_simps_2_3 = scipy.integrate.simps(poly(x_2_3), x_2_3)


#print results
print("Integral Values for 100 steps")
print("Quadrature\t|\tTrapezoidal\t|\tSimpson's")
print("{0:13.8f}{1:20.8f}{2:20.8f}\t\t Integral 1".format(integral_quad_1,integral_trapz_1,integral_simps_1))
print("{0:13.8f}{1:20.8f}{2:20.8f}\t\t Integral 2".format(integral_quad_2,integral_trapz_2,integral_simps_2))
print("\n")

print("Integral Values for 50 steps")
print("Quadrature\t|\tTrapezoidal\t|\tSimpson's")
print("{0:13.8f}{1:20.8f}{2:20.8f}\t\t Integral 1".format(integral_quad_1,integral_trapz_1_3,integral_simps_1_3))
print("{0:13.8f}{1:20.8f}{2:20.8f}\t\t Integral 2".format(integral_quad_2,integral_trapz_2_3,integral_simps_2_3))
print("\n")

print("Integral Values for 25 steps")
print("Quadrature\t|\tTrapezoidal\t|\tSimpson's")
print("{0:13.8f}{1:20.8f}{2:20.8f}\t\t Integral 1".format(integral_quad_1,integral_trapz_1_2,integral_simps_1_2))
print("{0:13.8f}{1:20.8f}{2:20.8f}\t\t Integral 2".format(integral_quad_2,integral_trapz_2_2,integral_simps_2_2))
print("\n")

print("""After playing around with the number of slices for the integral it appears that around 35 or less is 
when the integral begins to become very innacurate.""")
```
