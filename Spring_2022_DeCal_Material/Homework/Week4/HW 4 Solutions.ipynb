# Yilun's Random Walk in Campbell (Optional)

### This problem is just for your interest, we will revisit this later in the semester!

Consider a particle on a 2D xy-plane initially stays at the origin $\vec{r_0}=(0,0)$. In each time interval $\Delta t$, the particle would take one step with length $\delta$ to a RANDOM direction on the plane. Mathematically, the displacement vector for each step could be expressed as 

$$\Delta\vec{r} = (\Delta x, \Delta y) = (\delta\cos\theta, \delta\sin\theta) \quad \text{and} \quad |\Delta\vec{r}| = \delta\,\,,$$

where $\theta$ is defined to be the angle a vector makes with the positive x-axis. If this does not make sense to you, google "polar coordinates".

This is RANDOM WALK - it has many application in astronomy and physics, e.g. stellar interior, diffusion. In this problem, you are asked to plot the path of SEVEN randomly walking particles on a 2D plane after 10000 steps, all starting at the origin. Hint: `np.random.uniform` may be helpful. 

Your plot should look similar to the one in this link https://functions.wolfram.com/Constants/Pi/visualizations/2/02/imagetext/0044/text44.gif. 

On the plot, you should have a grid by using `plt.grid` (detailed instruction is on Google), and preferably, you mark out where the origin is. However, the axis labels are not important flr this plot, so you don't have to include them.

```python
# we provide the function for you
def random_walk(N):
    """
    A function that takes in the number of random 
    steps you want to take and returns an x and y 
    1D array corresponding to the path the particle 
    took in the xy plane. It also returns roughly 
    the limit of how far the particle can travel in 
    N steps.
    """
    N = int(N) #this is the number of steps 
    theta = np.random.uniform(0, 2*np.pi, size=N) #randomly choosing a direction theta
    x = np.cos(theta).cumsum(0)
    y = np.sin(theta).cumsum(0)
    average_lim = np.sqrt(N)
    return x,y,average_lim

### YOUR CODE HERE ###
# just plot 7 different runs of the random walk with the same N=10000

plt.figure(figsize=(12,8))
for i in range(7):
    x, y, lim = random_walk(N=1e6)
    plt.plot(x,y,lw=0.1, alpha=1)

plt.xlim((-lim*1.3,lim*1.3))
plt.ylim((-lim*1.3,lim*1.3))
plt.grid()

plt.show()
```
